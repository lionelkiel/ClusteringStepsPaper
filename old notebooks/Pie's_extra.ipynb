```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
dat1 = np.array([57,81,78,78])
dat2 = np.array([28,4,7,7])
data = np.array([dat2,dat1])
lab1 = "Number of states which proposed measures within this section of the value chain"
lab2 = "Number of states which did not propose measures within this section of the value chain"
```

```python
data.shape
```

```python
fig, ax = plt.subplots(1,4, figsize=(15,5))
# 4 pie charts
# font in times new roman
plt.rcParams['font.family'] = 'Times New Roman'
# font size
plt.rcParams['font.size'] = 12

pie = ax[0].pie(data[:,0], autopct='%1.0f%%', colors=['lightsteelblue', 'cornflowerblue'],  pctdistance=1.15, wedgeprops = {"edgecolor" :"white", 
                      'linewidth': 2, 
                      'antialiased': True}, startangle=90)
ax[1].pie(data[:,1], autopct='%1.0f%%', colors=['lightsteelblue', 'cornflowerblue'],  pctdistance=1.15, wedgeprops = {"edgecolor" :"white", 
                      'linewidth': 2, 
                      'antialiased': True}, startangle=90)
ax[2].pie(data[:,2], autopct='%1.0f%%', colors=['lightsteelblue', 'cornflowerblue'],  pctdistance=1.15, wedgeprops = {"edgecolor" :"white", 
                      'linewidth': 2, 
                      'antialiased': True}, startangle=90)
ax[3].pie(data[:,3], autopct='%1.0f%%', colors=['lightsteelblue', 'cornflowerblue'],  pctdistance=1.15, wedgeprops = {"edgecolor" :"white", 
                      'linewidth': 2, 
                      'antialiased': True}, startangle=90)

ax[0].set_title('Upstream', fontsize=16)
ax[1].set_title('Midstream', fontsize=16)
ax[2].set_title('Downstream', fontsize=16)
ax[3].set_title('Cross value chain', fontsize=16)

# add legend under the pie charts, no box around legend
ax[0].legend(labels=[lab1, lab2], loc='upper center', bbox_to_anchor=(1, -0.1), frameon=False)

plt.savefig('piecharts.pdf', bbox_inches='tight')
```
